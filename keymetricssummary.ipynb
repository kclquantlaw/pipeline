```python
import json
  
# Opening JSON file
f = open('housingact2004.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list

subsections={}

for i in data['sections']:
    print(i['section']['Number']+" "+str(i['subSectionsNum']))
    if i['subSectionsNum'] in subsections.keys():
        subsections[i['subSectionsNum']]=subsections[i['subSectionsNum']]+1;
    else:
        subsections[i['subSectionsNum']]=1

print(subsections)
maximum = max(subsections, key=subsections.get)  # Just use 'min' instead of 'max' for minimum.
print(maximum, subsections[maximum])

dict(sorted(subsections.items(), key=lambda item: item[1]))
```
