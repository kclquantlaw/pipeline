```python
import json
  
# Opening JSON file
f = open('housingact2004.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list

incoming={}

for i in data['sections']:
    
    print(i['section']['Number']+" "+str(i['outgoing']))
    
    for key in i['outgoing']:
        if key in incoming.keys():
             incoming[key]=i['outgoing'][key]+incoming[key];
        else:
            incoming[key]=i['outgoing'][key]
            
    print(i['section']['Number']+" "+str(i['ingoing']))


print(incoming)
print(len(incoming))

outgoing={}

for item in incoming:
    if("(" in item):
        keyName=item[0:item.index("(")-1]
        if keyName in outgoing.keys():
            outgoing[keyName]=incoming.get(item)+outgoing.get(keyName)
        else:
            outgoing[item[0:item.index("(")-1]]=incoming.get(item)
    else:
        if(outgoing.get(item) != None):
            outgoing[item]=incoming.get(item)+outgoing.get(item)
        else:
            outgoing[item]=incoming.get(item)            

            
print("New table")
print(len(outgoing))
print(dict(sorted(outgoing.items(), key=lambda item: item[1])))

# Closing file
f.close()
```
