```python
import json
  
# Opening JSON file
f = open('housingact2004.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
  
# Iterating through the json
# list

def clearSection(text):
    if(' ' in text):
        return text.replace(" ","")    
    elif('.' in text):
        return text.replace(".","")    
    else:
        return text;

incoming={}

for i in data['sections']:
        
    for key in i['ingoing']:
        if clearSection(key) in incoming.keys():
             incoming[clearSection(key)]=i['ingoing'][key]+incoming[clearSection(key)];
        else:
            incoming[clearSection(key)]=i['ingoing'][key]
        if(key=='52'): 
            print(i['section']['Number']+" "+str(i['ingoing']))

            
maximum = max(incoming, key=incoming.get)  # Just use 'min' instead of 'max' for minimum.
print(maximum, incoming[maximum])

print(incoming)

dict(sorted(incoming.items(), key=lambda item: item[1]))
```
