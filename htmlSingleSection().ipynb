```python
import json
    

def section(number):
    f = open('housingact2004.json')
    data = json.load(f)
    for i,key  in data['sections'][number]['section'].items():  
        if(i=='subSections'):
            for element in key:
                print(element['Title'])
                print(element['Number'])
                for paragraph in element['Paragraphs']:
                    print(paragraph['Title'])
                    print(paragraph['Number'])
                    for subParagraph in paragraph['subParagraphs']:
                        print(subParagraph['Title'])
                        print(subParagraph['Number'])
                        for line in subParagraph['Lines']:
                            print(line['Letter'])
                            print(line['Title'])
                            
section(0)

```

```python
def htmlRefLink(line, references):
    for item in references: 
        if('section '+str(item) in line or 'sections '+str(item)):
            line=line.replace(str(item), '<a href="#aside("'+str(item)+'")">'+str(item)+'</a>')
    return line

section_259_1="This section sets out when persons are to be treated for the purposes of section 254 as occupying a building or part of a building as their only or main residence."

print(htmlRefLink(section_259_1, [254]))

# a to reference

section_1="in multiple occupation as defined by sections 254 to 259, as they have effect for the purposes of this Part (that is, without the exclusions contained in Schedule 14)"

print(htmlRefLink(section_1, [254, 259]))
```
